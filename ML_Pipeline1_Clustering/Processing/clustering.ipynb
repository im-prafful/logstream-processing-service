```python
import pandas as pd
from sqlalchemy import create_engine

DB_USER = "masterUser"
DB_PASS = "Admin$1234"
DB_NAME = "Logstream_DB"

# Since SSH tunnel is already active:
engine = create_engine(
    f"postgresql+psycopg2://{DB_USER}:{DB_PASS}@localhost:5433/{DB_NAME}"
)

# Read entire table into DataFrame
df = pd.read_sql("SELECT * FROM logs;", engine)

print(df.head())
```

```python
df = df[["level", "message", "source", "parsed_data"]]
print(df.columns.tolist())
```

```python
df["level_encoded"] = df["level"].map({"error": 0, "warning": 1})# encode levels cols..

#freq_encode source col
freq_map = df["source"].value_counts().to_dict()
df["source_encoded"] = df["source"].map(freq_map)

print(df)
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

model=SentenceTransformer('all-MiniLM-L6-v2')

embeddings=np.vstack(df['message'].apply(lambda x:model.encode(x)))

print(embeddings.shape)


```
